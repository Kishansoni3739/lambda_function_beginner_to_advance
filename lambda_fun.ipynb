# Lambda Function Basics - Practice Through Questions

### Q1. - Write a lambda to add two numbers.

```python
sum = lambda a,b : a + b
print(sum(3,8))
```

### Q2. - Lambda to square a number.

```python
square  = lambda a : a**2
print(square(4))
```

### Q3. - Lambda to check if a number is even or odd.

```python
check_even = lambda x : "Even" if x % 2 == 0 else "Odd"
print(check_even(2))
```

### Q4. - Lambda to return the larger of two numbers.

```python
check_large = lambda x, y : x if x > y else y
print(check_large(4,8))
```

### Q5. - Lambda to convert Celsius to Fahrenheit.

```python
to_fahrenheit = lambda x : (x * 9/5) + 32
print(to_fahrenheit(38))
```

### Q6. - Given [1,2,3,4,5], use lambda + map() to square every number.

```python
l = [1,2,3,4,5]
square_l = list(map(lambda x : x**2, l))
print(square_l)
```

### Q7. - Convert list of strings to uppercase using lambda.

```python
name = "kishan"
upper_name = lambda x : x.upper()
print(upper_name(name))
```

### Q8. - From [10,20,30], create a new list adding 5 to each element.

```python
l2 = [10, 20, 30]
add_five = list(map(lambda x : x+5, l2))
print(add_five)
```

### Q9. - Given list of prices, apply 10% discount using lambda.

```python
prices = [235,600, 850]
discounted_price = list(map(lambda x : x - (x/10), prices))
print(discounted_price)
```

### Q10. - Convert list of integers to strings using lambda.

```python
integers = [3,4,8,23,46]
to_string  = list(map(lambda x : str(x), integers))
print(to_string)
```

### Q11. - From [1..20], filter even numbers.

```python
l4 = [x for x in range(1, 21)]
even_only = list(filter(lambda x : x % 2 == 0, l4))
even_only
```

### Q12. - Filter numbers greater than 50 from list.

```python
l5 = [43, 67, 83, 13, 90, 56]
larger_than_fifty = list(filter(lambda x : x > 50, l5))
larger_than_fifty
```

### Q13. - Filter words longer than 4 characters.

```python
names = ['lokesh','kishan','raju','om','saran']
remain_names = list(filter(lambda x : len(x) > 4, names))
remain_names
```

### Q14. - From list of names, filter names starting with 'A'.

```python
names2 = ['lokesh','ankita', 'praveen','aman','yash','Ankit']
names_with_a = list(filter(lambda x : x if x[0].lower() == 'a' else None, names2))
names_with_a
```

### Q15. - Sort list of tuples by second element

```python
data = [(1,3),(2,1),(4,2)]
sorted_data = sorted(data, key = lambda x : x[1])
sorted_data
```

### Q16. - Sort list of strings by length.

```python
s1 = ['tarun','kishan','lokesh','raju','om']
sorted_s1 = sorted(s1, key = lambda x : len(x))
sorted_s1
```

### Q17. - Sort students list by marks descending

```python
students=[("Ram",78),("Shyam",45),("Aman",90)]
sorted_students = sorted(students, key = lambda x : x[1], reverse = True)
sorted_students
```

### Q18. - Use lambda to multiply all numbers in a list.

```python
from functools import reduce
l6 = [1,2,3,4,5]
multiplied = reduce(lambda a,b : a*b, l6)
multiplied
```

### Q19. - Find maximum number in list using lambda.

```python
l7 = [15,89,45,93,235,5,89,33]
largest = reduce(lambda a,b : a if a > b else b, l7)
largest
```

### Q20. - Write lambda to reverse a string.

```python
text = 'lokesh'
reverse_text = lambda x : x[::-1]
reverse_text(text)
```

### Q21. - Lambda to check palindrome.

```python
text_list = ['naman','raju','nitin','komal']
palindrome_text = list(filter(lambda x : x if x[::-1] == x else None, text_list))
palindrome_text
```
